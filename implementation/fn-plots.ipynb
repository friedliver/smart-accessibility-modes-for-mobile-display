```python
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Emulate Dark Mode
x = np.linspace(start=0, stop=255, num=255)
y = np.linspace(start=0, stop=255, num=255)
for i in range(255):
    y[i] = 255 if x[i] == 0 else min(255, 1000000 / math.pow(x[i], 2))
plt.figure(figsize = (6, 4))
plt.plot(x, y)
plt.title('Emulate Dark Mode L-value Mapping')
plt.xlabel('Original L-value')
plt.ylabel('Target L-value')
plt.show()
```

```python
# Improve Contrast
x = np.linspace(start=0, stop=255, num=255)
y = np.linspace(start=0, stop=255, num=255)
for i in range(255):
    y[i] = 208 / (1 + (math.pow(math.e, (40 - x[i]) / 4)))
plt.plot(x, y)
plt.title('Improve Contrast L-value Mapping')
plt.xlabel('Original L-value')
plt.ylabel('Target L-value')
plt.show()
```
