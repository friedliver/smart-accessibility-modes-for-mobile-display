```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def detect_edge(image):
    ''' Detecting Edges '''
    image_with_edges = cv2.Canny(image , 10, 100)
    images = [image , image_with_edges]
    location = [121, 122]
    plt.figure(figsize = (20, 20))
    for loc, img in zip(location, images):
        plt.subplot(loc)
        plt.imshow(img, cmap='gray')
    plt.show()
    return image_with_edges
```

```python
image = cv2.imread('../screens/01_bank-app.jpg', 0)
detect_edge(image)
```

```python
image = cv2.imread('../screens/02_youtube-chat.jpg', 0)
detect_edge(image)
```
